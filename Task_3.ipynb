```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics 
import warnings
warnings.filterwarnings("ignore")
```

__Givens__

```python
Background=np.array([76.9,91,114.2,154.2,88.1,18.6,11.4,31.1,69.6,105.7,112.7,77.4])
Observations=np.array([80,90,110,160,80,20,10,40,70,100,110,80])
```

```python
H=np.eye(12)
B=0.1*np.eye(12)
R=0.25*np.eye(12)
x=Background.T
y=Observations.T
```

### (a) Computational of an optimal estimates for rainfall

```python
def Optimal_estimates(x,y,H,B,R):
    K=(B@H.T)*np.linalg.inv(H@B@H.T+R)
    xh=x+K@(y-H@x)
    return xh


Result=Optimal_estimates(x,y,H,B,R)
print(Result)
```

#### Corresponding plot

```python
t=np.arange(1,13,1)
plt.figure(figsize=(8,6))
plt.plot(t,Result, 'o',label = 'Opt_estimates')
plt.plot(t,x, 'o',label = 'Background')
plt.plot(t,y, 'o',label = 'Data')
plt.title("Rainfall in Kigali-2019")
plt.legend()
plt.ylabel('Rainfall')
plt.xlabel('Months')
plt.grid()
plt.show()
```

__From the graph we can observe that even if the background are trying to be closer with the optimal estimates monthly rainfall, the rainfall obesvations are the closest to optimal compare with the background. Simply, the almost year the optimal estimates are fitting the rainfall observations.__

### (b) Computational of an optimal estimates for rainfall

```python
B_b=0.25*np.eye(12)
R_b=0.1*np.eye(12)
Result_b=Optimal_estimates(x,y,H,B_b,R_b)
print(Result_b)
```

#### Corresponding plot

```python
plt.figure(figsize=(8,6))
plt.plot(t,Result_b, 'o',label = 'Opt_estimates')
plt.plot(t,x, 'o',label = 'Background')
plt.plot(t,y, 'o',label = 'Data')
plt.title("Rainfall in Kigali-2019")
plt.legend()
plt.xlabel('Months')
plt.ylabel('Rainfall')
plt.grid()
plt.show()
```

__The plot after to change the background error variance from $0.1$ to $0.25 mm$, and observation error variance from $0.25$ to $0.1 mm$, the plot is almost the same as the like the previous. This is to mean that the observation data for rainfall are still the closest to the optimal estimates compare to the background.__

### (c) Computational of an optimal estimates for rainfall

```python
H[5:8,5:8]=0
H_c=H
B_c=0.25*np.eye(12)
R_c=0.25*np.eye(12)
Result_c=Optimal_estimates(x,y,H_c,B_c,R_c)
print(Result_c)
```
